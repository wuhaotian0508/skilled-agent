```python
import yaml
import os

skill_path = '/data/haotianwu/simple-agent/skills/dev-assistant/SKILL.md'

with open(skill_path, "r", encoding="utf-8") as f:
    content = f.read()

# 自动从路径中提取文件夹名
name = os.path.basename(os.path.dirname(skill_path))  # 得到 'dev-assistant'

skill_xml = ""
parts = content.split("---", 2)

if len(parts) >= 3:
    meta = yaml.safe_load(parts[1])
    skill_xml += f"""
<skill>
<name>{meta.get('name', name)}</name>
<description>{meta.get('description', '')}</description>
<path>{skill_path}</path>
</skill>"""
    
    print("生成的 XML:")
    print(skill_xml)
    print("\n解析的元数据:")
    print(meta)
else:
    print("文件格式不正确，缺少 YAML front matter")
```
